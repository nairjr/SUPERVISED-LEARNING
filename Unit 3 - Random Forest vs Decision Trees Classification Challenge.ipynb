```python


# Packages for rendering our tree.
import pydotplus
import graphviz

# A convenience for displaying visualizations.
from IPython.display import Image


# Initialize and train our tree.
decision_tree = tree.DecisionTreeClassifier(
    criterion='entropy',
    max_features=1,
    max_depth=3,
    random_state = 1337 #this makes the result reproducible 
)
decision_tree.fit(X,Y)

# Render our tree.
dot_data = tree.export_graphviz(
    decision_tree, out_file=None,
    feature_names=feature_names,
    class_names = ['1 - Mammal', '2 - Bird', '3 - Reptile', '4 - Fish', '5 - Amphibian', '6 - Bug', '7 - Invertebrate'],
    #class_names=zoo['class_type'].apply(str), # should this label every single row? should it specify the labels that we're using?
    # now it's working as labeling everying single row
    filled=True
)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

### Simplicity via Runtime

```python
import time

#runtime of tree
start = time.time()

regressor_tree = tree.DecisionTreeClassifier(
    criterion = 'entropy',
    max_depth = 3,
    max_features = 3
)

tree_Score = cross_val_score(regressor_tree, X, Y, cv=5)

end = time.time()
tree_time = end-start
print(tree_time)
```

```python

#runtime of forest
start = time.time()

regressor_forest = ensemble.RandomForestRegressor(
    max_depth = 3,
    max_features = 3
)

forest_score = cross_val_score(regressor_forest, X, Y, cv=5)

end = time.time()
forest_time = end-start
print(forest_time)
```

```python
print('Forest is', round(forest_time/tree_time, 2), 'times slower than the tree')
```
