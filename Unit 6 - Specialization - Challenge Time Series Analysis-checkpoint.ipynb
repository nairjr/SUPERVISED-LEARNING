```python
import numpy as np
import pandas as pd
import scipy
from datetime import datetime
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nairjr/Datasets/master/A191RL1Q225SBEA.csv')
```

```python
df = df.rename(columns={'A191RL1Q225SBEA': 'GDP'})
```

```python
df.head()
```

```python
df.index = pd.to_datetime(df.DATE)
df.plot()
plt.title('GDP change')
plt.show()
```

```python
df['diff_1'] = df.GDP - df.GDP.shift()
df.plot(y='diff_1')
plt.title('GDP Time Series')
plt.show()
```

```python
#pacf
from statsmodels.tsa.stattools import pacf
x_acf = pd.DataFrame(pacf(df.GDP))
x_acf.plot(kind='bar')
plt.title("GDP PACF")
```

```python
x_acf = pd.DataFrame(pacf(df.diff_1[1:]))
x_acf.plot(kind='bar')
plt.title("GDP Diff PACF")
plt.legend(['difference'])
```

## Building models.

```python
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(df.GDP, order=(0,1,0))
model_fit = model.fit()
print(model_fit.summary())
print('Residuals Description')
print(model_fit.resid.describe())
```

```python
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(df.GDP, order=(1,0,0))
model_fit = model.fit()
print(model_fit.summary())
print('Residuals Description')
print(model_fit.resid.describe())
```

```python
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(df.GDP, order=(1,0,1))
model_fit = model.fit(disp=0)
print(model_fit.summary())
print('Residuals Description')
print(model_fit.resid.describe())
```

```python
from statsmodels.tsa.arima_model import ARIMA
model_111 = ARIMA(df.GDP, order=(1,1,1))
model_111_fit = model_111.fit()
print(model_111_fit.summary())
print('Residuals Description')
print(model_111_fit.resid.describe())
```

```python
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(df.GDP, order=(5,1,5))
model_fit = model.fit()
print(model_fit.summary())
print('Residuals Description')
print(model_fit.resid.describe())
```

## Residuals

```python
residuals = pd.DataFrame(model_111_fit.resid)

residuals.plot(legend=False)
plt.title('Time Series of Residuals')

residuals.hist(bins=20)
plt.title('Histogram of Residuals')
```
